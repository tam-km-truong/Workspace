```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#bakrep batches
path = './new_bakrep_batches'
df_cpr_size = pd.read_csv(path+"/batch_cpr_sizes.csv", names = ['size', 'batch_name'], index_col=['batch_name'])
df_genome_counts = pd.read_csv(path+"/genome_counts.csv", names = ['count', 'batch_name'], index_col=['batch_name'])
df_kmers_counts = pd.read_csv(path+"/distinct_kmers_counts.csv", names = ['batch_name', 'kmers_count'], index_col=['batch_name'])
# merge 3 of them
df_bakrep_all = pd.concat([df_cpr_size, df_genome_counts, df_kmers_counts], axis=1)
```

```python
df_bakrep_all
```

```python
#bakrep batches HQ
path = './bakrep_HQ'
df_cpr_size = pd.read_csv(path+"/batch_cpr_sizes.csv", names = ['size', 'batch_name'], index_col=['batch_name'])
df_genome_counts = pd.read_csv(path+"/genome_counts.csv", names = ['count', 'batch_name'], index_col=['batch_name'])
df_kmers_counts = pd.read_csv(path+"/distinct_kmers_counts.csv", names = ['batch_name', 'kmers_count'], index_col=['batch_name'])
# merge 3 of them
df_bakrep_HQ = pd.concat([df_cpr_size, df_genome_counts, df_kmers_counts], axis=1)
```

```python
df_bakrep_HQ
```

```python
# byte per genomes
df_bakrep_HQ["byte_per_genome"] = df_bakrep_HQ["size"] / df_bakrep_HQ["count"]
df_bakrep_all["byte_per_genome"] = df_bakrep_all["size"] / df_bakrep_all["count"]
```

```python
# bit per kmers
df_bakrep_HQ["bit_per_kmer"] = (df_bakrep_HQ["size"] * 8) / df_bakrep_HQ["kmers_count"]
df_bakrep_all["bit_per_kmer"] = (df_bakrep_all["size"] * 8) / df_bakrep_all["kmers_count"]
```

```python
df_bakrep_HQ
```

```python
df_bakrep_all
```

```python
df = df_bakrep_all
df["kilobyte_per_genome"] = df["byte_per_genome"] / 1024

# Create mask
dustbin_mask = df.index.str.startswith("dustbin")

# Normalize count for point size (e.g., scale 4000 → size 200)
max_size = 200
df["point_size"] = (df["count"] / df["count"].max()) * max_size

# Create the plot
plt.figure(figsize=(8, 6))

# Plot dustbin points in black
plt.scatter(
    df.loc[dustbin_mask, "bit_per_kmer"],
    df.loc[dustbin_mask, "kilobyte_per_genome"],
    s=df.loc[dustbin_mask, "point_size"],
    color="darkgrey", label="dustbin", edgecolors='k'
)

# Plot other points in blue
plt.scatter(
    df.loc[~dustbin_mask, "bit_per_kmer"],
    df.loc[~dustbin_mask, "kilobyte_per_genome"],
    s=df.loc[~dustbin_mask, "point_size"],
    color="blue", label="other", edgecolors='k'
)

# Labels and title
plt.xlabel("Bits per distinct k-mer")
plt.ylabel("Kilobytes per genome")
plt.title("Compression Efficiency per Batch - All batches")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df = df_bakrep_HQ
df["kilobyte_per_genome"] = df["byte_per_genome"] / 1024

# Create mask
dustbin_mask = df.index.str.startswith("dustbin")

# Normalize count for point size (e.g., scale 4000 → size 200)
max_size = 200
df["point_size"] = (df["count"] / df["count"].max()) * max_size

# Create the plot
plt.figure(figsize=(8, 6))

# Plot dustbin points in black
plt.scatter(
    df.loc[dustbin_mask, "bit_per_kmer"],
    df.loc[dustbin_mask, "kilobyte_per_genome"],
    s=df.loc[dustbin_mask, "point_size"],
    color="red", label="dustbin", edgecolors='k'
)

# Plot other points in blue
plt.scatter(
    df.loc[~dustbin_mask, "bit_per_kmer"],
    df.loc[~dustbin_mask, "kilobyte_per_genome"],
    s=df.loc[~dustbin_mask, "point_size"],
    color="green", label="other", edgecolors='k'
)

# Labels and title
plt.xlabel("Bits per distinct k-mer")
plt.ylabel("Kilobytes per genome")
plt.title("Compression Efficiency per Batch - HQ")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# plot the 2 at the same time
##df_1
df_1 = df_bakrep_all
df_1["kilobyte_per_genome"] = df_1["byte_per_genome"] / 1000

dustbin_mask = df_1.index.str.startswith("dustbin")

max_size = 150
df_1["point_size"] = (df_1["count"] / df_1["count"].max()) * max_size

# Create the plot
plt.figure(figsize=(10, 8))

plt.scatter(
    df_1.loc[dustbin_mask, "bit_per_kmer"],
    df_1.loc[dustbin_mask, "kilobyte_per_genome"],
    s=df_1.loc[dustbin_mask, "point_size"],
    color="darkgrey", label="dustbin - All", edgecolors='k'
)

plt.scatter(
    df_1.loc[~dustbin_mask, "bit_per_kmer"],
    df_1.loc[~dustbin_mask, "kilobyte_per_genome"],
    s=df_1.loc[~dustbin_mask, "point_size"],
    color="blue", label="other - All", edgecolors='k'
)

## df_2

df_2 = df_bakrep_HQ
df_2["kilobyte_per_genome"] = df_2["byte_per_genome"] / 1000

dustbin_mask = df_2.index.str.startswith("dustbin")

df_2["point_size"] = (df_2["count"] / df_2["count"].max()) * max_size

plt.scatter(
    df_2.loc[dustbin_mask, "bit_per_kmer"],
    df_2.loc[dustbin_mask, "kilobyte_per_genome"],
    s=df_2.loc[dustbin_mask, "point_size"],
    color="red", label="dustbin - HQ", edgecolors='k'
)

plt.scatter(
    df_2.loc[~dustbin_mask, "bit_per_kmer"],
    df_2.loc[~dustbin_mask, "kilobyte_per_genome"],
    s=df_2.loc[~dustbin_mask, "point_size"],
    color="green", label="other - HQ", edgecolors='k'
)

# Labels and title
plt.xlabel("Bits per distinct k-mer")
plt.ylabel("Kilobytes per genome")
plt.title("Compression Efficiency per Batch - Combined")
plt.grid(True)
plt.xticks(np.arange(0, 26, 1))
plt.yticks(np.arange(0, 675, 25))
plt.tight_layout()
plt.legend()
plt.show()
```
