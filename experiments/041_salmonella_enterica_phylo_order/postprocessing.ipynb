```python
import pandas as pd
import numpy as np
import os
```

```python
path_index = 'remaining_genomes.txt'
indexes = np.loadtxt(path_index, delimiter=",",dtype='str')
indexes = indexes[:]
indexes
```

```python
values = []
for i in range(10):
    values = values + [f'min_values/distance.part_00{i}.txt']
```

```python
values
```

```python
values_to_numpy = {}
for i in range(10):
    values_to_numpy[i] = np.loadtxt(values[i], delimiter=",",dtype='str')[:]
```

```python
values_to_numpy[1]
```

```python
df_values = pd.DataFrame()
df_values.index = indexes
```

```python
for i in range(10):
    df_values[f'file_{i}'] = values_to_numpy[i]
```

```python
df_values
```

```python
np_coord = np.array(df_values.idxmin(axis=1))
np_coord
```

```python
names = []
for i in range(10):
    names = names + [f'min_names/distance.part_00{i}.txt']
```

```python
names
```

```python
names_to_numpy = {}
for i in range(10):
    names_to_numpy[i] = np.loadtxt(names[i], delimiter=",",dtype='str')[:]
```

```python
names_to_numpy[1]
```

```python
df_names = pd.DataFrame()
df_names.index = indexes
```

```python
for i in range(10):
    df_names[f'file_{i}'] = names_to_numpy[i]
```

```python
df_names
```

```python
rows = indexes
cols = np_coord
```

```python
# Extract values row by row
result_values = [df_names.at[row, col] for row, col in zip(rows, cols)]

# Create the result DataFrame
df_res = pd.DataFrame(result_values, index=rows, columns=['result'])
```

```python
df_res
```

```python
sizes = df_res.groupby('result').size()
sizes
```

```python
sizes.max()
```

```python
references = np.loadtxt('references.txt', delimiter=",",dtype='str')
```

```python
references = pd.Series(0, index=references)
references
```

```python
references.update(sizes)
references
```

```python
groups = df_res.groupby('result').groups
```

```python
len(groups['SAMN02766947'])
```

```python
path_to_save = 'phylo_orders/'
if not os.path.exists(path_to_save):
    os.mkdir(path_to_save)
```

```python
filename = f'{path_to_save}/phylo_order.txt'
with open(filename, "w") as f:
    for key in references.index:
        print(key)
        if key in groups:
            genomes = [key]+list(groups[key])
            for gen in genomes:
                f.write(f"{gen}\n")
        else:
            f.write(f"{key}\n")
        # values = [key]+list(grby_groups[key])
        # write_list_to_file(f'{path_to_save}/{key}.txt',values)
```
