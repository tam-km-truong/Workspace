```python
path_to_data = '/Users/ktruong/workspace/Workspace/experiments/045_selection_strategy_experiments/mash_distances/'
dis_25 = pd.read_csv(path_to_data+'distance_25.csv', header=0,index_col=0,on_bad_lines='warn',low_memory=False)
dis_50 = pd.read_csv(path_to_data+'distance_50.csv', header=0,index_col=0,on_bad_lines='warn',low_memory=False)
dis_75 = pd.read_csv(path_to_data+'distance_75.csv', header=0,index_col=0,on_bad_lines='warn',low_memory=False)
dis_100 = pd.read_csv(path_to_data+'distance_100.csv', header=0,index_col=0,on_bad_lines='warn',low_memory=False)
```

```python
print(dis_25.shape)
dis_25.head()
```

```python
idx_min_25 = dis_25.idxmin(axis = 1)
idx_min_25
```

```python
grby_size = idx_min_25.groupby(idx_min_25).size()
grby_size.sort_values()
```

```python
references_25 = pd.Series(0, index=dis_25.columns)
references_25
```

```python
references_25.update(grby_size)
```

```python
references_25.max(),references_25.mean()
```

```python
grby_groups = idx_min_25.groupby(idx_min_25).groups
grby_groups
```

```python
first_order_25 = []
i=1
for key in references_25.index:
    print(i)
    i+=1
    if key in grby_groups:
        values = [key]+list(grby_groups[key])
        first_order_25 = first_order_25 + values
    else:
        first_order_25 = first_order_25 + [key]
```
