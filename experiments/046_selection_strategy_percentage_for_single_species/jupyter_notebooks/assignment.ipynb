```python
idx_min_01 = dis_01.idxmin(axis = 1)
idx_min_01
```

```python
sizes = idx_min_01.groupby(idx_min_01).size()
sizes.sort_values()
```

```python
groups = idx_min_01.groupby(idx_min_01).groups
groups
```

```python
references_01 = pd.Series(0, index=dis_01.columns)
references_01
```

```python
references_01.update(sizes)
references_01.max(),references_01.mean()
```

```python
first_order_01 = []
for key in references_01.index:
    if key in groups:
        values = [key]+list(groups[key])
        first_order_01 = first_order_01 + values
    else:
        first_order_01 = first_order_01 + [key]
```
