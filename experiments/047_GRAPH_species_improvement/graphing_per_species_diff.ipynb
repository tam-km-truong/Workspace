```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
genome_counts = pd.read_csv('genome_counts.csv', header=None, names=['count', 'batch_id'], index_col='batch_id')
genome_counts
```

```python
new_strat = pd.read_csv('new_strat.csv', header=None, names=['size_new_(byte)', 'batch_id'], index_col='batch_id')
new_strat
```

```python
old_strat = pd.read_csv('old_strat.csv', header=None, names=['size_old_(byte)', 'batch_id'], index_col='batch_id')
old_strat
```

```python
compression_sizes = new_strat
```

```python
compression_sizes['size_old_(byte)'] = old_strat['size_old_(byte)'].values
compression_sizes['count'] = genome_counts['count'].values
```

```python
compression_sizes
```

```python
species = [s[:-11] for s in compression_sizes.index]
species
```

```python
compression_sizes['species'] = species
compression_sizes
```

```python
size_by_species =  compression_sizes.groupby('species').sum().sort_values(by='size_old_(byte)')
size_by_species
```

```python
size_by_species['diff'] = (size_by_species['size_old_(byte)'] - size_by_species['size_new_(byte)'])
size_by_species['percentage'] = round((size_by_species['diff'] / size_by_species['size_old_(byte)'])*100,2)
```

```python
species_more_4k  =  size_by_species[size_by_species['count']>=4000].sort_values(by='count', ascending=False)
species_more_4k
```

```python
species_more_4k.mean()
```

```python
fig, ax = plt.subplots(figsize=(14, 8))

df_plot = species_more_4k.loc[:,'percentage']

# '$\it{text you want to show in italics}$'
files = [f"{' '.join(f.split('_'))}" for f in df_plot.index.values]
counts = species_more_4k['count'].values
files = [f"{files[i]}\n{counts[i]}"for i in range(len(files))]

sizes = df_plot.values
# bar_labels = ['red', 'blue', '_red', 'orange']
# bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

# ax.bar(files, counts, label=bar_labels, color=bar_colors)
ax.bar(files, sizes, edgecolor='black')

ax.set_ylabel('%',fontsize=16)
ax.set_title('Result: Compression Size Reduction Per Species',fontsize=18, fontweight='bold')
ax.set_xlabel('Species with more than 4000 genomes',fontsize=16)
#ax.invert_yaxis()
#ax.xaxis.tick_top()

plt.axhline(y=np.nanmean(sizes), color = 'navy', linestyle='--')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

plt.yticks(np.arange(0, 31, 1))
plt.xticks(rotation=80,ha='center',fontstyle = 'italic')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 8))

df_plot = species_more_4k.loc[:,'diff']

# '$\it{text you want to show in italics}$'
files = [f"{' '.join(f.split('_'))}" for f in df_plot.index.values]
counts = species_more_4k['count'].values
files = [f"{files[i]}\n{counts[i]}"for i in range(len(files))]

sizes = df_plot.values / 1000000000
# bar_labels = ['red', 'blue', '_red', 'orange']
# bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

# ax.bar(files, counts, label=bar_labels, color=bar_colors)
ax.bar(files, sizes, edgecolor='black', color='lightgreen')

ax.set_ylabel('Size reduction (GB)',fontsize=16)
ax.set_title('Result: Compression Size Reduction Per Species',fontsize=18, fontweight='bold')
ax.set_xlabel('Species with more than 4000 genomes',fontsize=16)
#ax.invert_yaxis()
#ax.xaxis.tick_top()

#plt.axhline(y=np.nanmean(sizes), color = 'navy', linestyle='--')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
plt.xticks(rotation=80,ha='center',fontstyle = 'italic')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 8))

df_plot = species_more_4k.loc[:,'diff']

# '$\it{text you want to show in italics}$'
files = [f"{' '.join(f.split('_'))}" for f in df_plot.index.values]
counts = species_more_4k['count'].values
files = [f"{files[i]}\n{counts[i]}"for i in range(len(files))]

sizes = df_plot.values / 1000000000
above_diff = df_plot.values / 1000000000
below_old_size = species_more_4k.loc[:,'size_old_(byte)'].values - species_more_4k.loc[:,'diff'].values
below_old_size = below_old_size / 1000000000
weight_counts = {
    "New compression size": below_old_size,
    "Difference": above_diff,
}
# bar_labels = ['red', 'blue', '_red', 'orange']
# bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

# ax.bar(files, counts, label=bar_labels, color=bar_colors)
bottom = np.zeros(20)
for boolean, weight_count in weight_counts.items():
    p = ax.bar(files, weight_count, label=boolean, bottom=bottom, edgecolor =  'black')
    bottom += weight_count

ax.set_ylabel('Compression Size (GB)',fontsize=16)
ax.set_title('Result: Compression Size Reduction Per Species',fontsize=18, fontweight='bold')
ax.set_xlabel('Species with more than 4000 genomes',fontsize=16)
#ax.invert_yaxis()
#ax.xaxis.tick_top()

#plt.axhline(y=np.nanmean(sizes), color = 'navy', linestyle='--')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
plt.xticks(rotation=80,ha='center',fontstyle = 'italic')
ax.legend(loc="upper right")
plt.show()
```
