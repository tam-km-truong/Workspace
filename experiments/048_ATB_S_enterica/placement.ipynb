```python
import pandas as pd
import numpy as np
import os
```

```python
path_index = 'remains_s_enterica.csv'
```

```python
indexes = np.loadtxt(path_index, delimiter=",",dtype='str')
indexes = indexes[:-1]
indexes,len(indexes)
```

```python
values_01 = 'values/values_01.csv'
values_02 = 'values/values_02.csv'
```

```python
np_val_1 = np.loadtxt(values_01, delimiter=",",dtype='str')
np_val_2 = np.loadtxt(values_02, delimiter=",",dtype='str')
```

```python
df_values = pd.DataFrame()
df_values.index = indexes
df_values['file_1'] = np_val_1
df_values['file_2'] = np_val_2
```

```python
df_values
```

```python
np_coord = np.array(df_values.idxmin(axis=1))
np_coord
```

```python
names_1 = 'names/names_01.csv'
names_2 = 'names/names_02.csv'
```

```python
np_names_1 = np.loadtxt(names_1, delimiter=",",dtype='str')
np_names_2 = np.loadtxt(names_2, delimiter=",",dtype='str')
```

```python
df_names = pd.DataFrame()
df_names.index = indexes
df_names['file_1'] = np_names_1
df_names['file_2'] = np_names_2
```

```python
df_names
```

```python
rows = indexes
cols = np_coord
```

```python
# Extract values row by row
result_values = [df_names.at[row, col] for row, col in zip(rows, cols)]

# Create the result DataFrame
df_res = pd.DataFrame(result_values, index=rows, columns=['result'])
```

```python
df_res
```

```python
sizes = df_res.groupby('result').size()
sizes
```

```python
sizes.idxmax(),sizes.max()
```

```python
references = np.loadtxt('references/leaf_order.txt', delimiter=",",dtype='str')
```

```python
references
```

```python
references = pd.Series(0, index=references)
references
```

```python
references.update(sizes)
references
```

```python
groups = df_res.groupby('result').groups
```

```python
path_to_save = 'phylo_orders/'
if not os.path.exists(path_to_save):
    os.mkdir(path_to_save)
```

```python
filename = f'{path_to_save}/groupby_order.txt'
with open(filename, "w") as f:
    for key in references.index:
        print(key)
        if key in groups:
            values = [key]+list(groups[key])
            for value in values:
                f.write(f"{value}\n")
        else:
            f.write(f"{key}\n")
```
