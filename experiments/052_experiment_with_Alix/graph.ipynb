```python
size_miniphy_ecoli = 232724440
size_vp_ecoli = 237786552

size_miniphy_sal_enterica_12k = 289166948
size_vp_sal_enterica_12k = 292352064
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
names = ['Phylo','TSP',]
counts = [size_miniphy_sal_enterica_12k, size_vp_sal_enterica_12k]

bar_container = ax.bar(names, counts, edgecolor = 'black', color='lightblue')
#ax.set(ylim=(0, 3.5))
ax.set_ylabel('Compression size (MB)',fontsize=16)
#ax.set_title('Result: Compression Size Of 661k Collection',fontsize=18, fontweight='bold')
ax.set_xlabel('Method',fontsize=16)
ax.ticklabel_format(axis = 'y',scilimits = (6,6))
ax.bar_label(bar_container, labels = [round(c/1000000,2) for c in counts])
ax.set(ylim=(0, 300000000))


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
```
