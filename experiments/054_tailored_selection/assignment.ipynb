```python
import pandas as pd
import numpy as np
import os
```

```python
dis = pd.read_csv('./distance.csv', header=0,index_col=0,on_bad_lines='warn',low_memory=False)
```

```python
print(dis.shape)
dis.head()
```

```python
idx_min = dis.idxmin(axis = 1)
idx_min
```

```python
sizes = idx_min.groupby(idx_min).size()
sizes.sort_values()
```

```python
groups = idx_min.groupby(idx_min).groups
groups
```

```python
references = pd.Series(0, index=dis.columns)
references
```

```python
references.update(sizes)
references.max()
```

```python
references
```

```python
first_order = []
for key in references.index:
    if key in groups:
        values = [key]+list(groups[key])
        first_order = first_order + values
    else:
        first_order = first_order + [key]
```

```python
len(first_order)
```
