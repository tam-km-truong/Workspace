```python
import subprocess
from collections import defaultdict

# === Config ===
DISTANCE_FILE = distance_file
ROW_NAMES_FILE = "./references_selection/e_coli/0.1/Escherichia_coli_661k_remaining_0.1_percent.txt"
COL_NAMES_FILE = "./references_selection/e_coli/0.1/Escherichia_coli_661k_reference_0.1_percent.txt"

# === Step 1: Read row and column names ===
with open(ROW_NAMES_FILE) as f:
    row_names = [line.strip() for line in f]

with open(COL_NAMES_FILE) as f:
    col_names = [line.strip() for line in f]

# === Step 2: Run awk to get argmin index per row ===
awk_script = r'''
{
    min = $1; idx = 0;
    for (i = 2; i <= NF; i++) {
        if ($i < min) {
            min = $i;
            idx = i;
        }
    }
    print idx
}
'''

result = subprocess.run(
    ["awk", awk_script],
    stdin=open(DISTANCE_FILE),
    capture_output=True,
    text=True,
    check=True
)

argmin_indices = [int(i) for i in result.stdout.strip().split("\n")]

# === Step 3: Group rows by argmin column ===
groups = defaultdict(list)

for row_name, idx in zip(row_names, argmin_indices):
    col_name = col_names[idx]
    groups[col_name].append(row_name)

# === Step 4: Output ===
for col, grouped_rows in groups.items():
    print(f"{col}:\t{grouped_rows}")
```
