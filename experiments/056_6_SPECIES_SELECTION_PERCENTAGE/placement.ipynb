```python
import pandas as pd
import numpy as np
import os
```

```python
path_to_data = '/Users/ktruong/workspace/Workspace/experiments/056_6_SPECIES_SELECTION_PERCENTAGE/distances/'
references = np.loadtxt('/Users/ktruong/workspace/Workspace/experiments/056_6_SPECIES_SELECTION_PERCENTAGE/skeleton_trees/e_coli/leaf_order_01.txt', dtype=str)
```

```python
dis = pd.read_csv(path_to_data+'e_coli/0.1/distances.tsv',sep='\t', header=0,index_col=0,on_bad_lines='warn',low_memory=False)
```

```python
dis
```

```python
idx = [os.path.basename(i)[:-6] for i in dis.index]
```

```python
cols =[os.path.basename(i)[:-6] for i in dis.columns]
```

```python
dis.index = idx
dis.columns = cols
```

```python
dis
```

```python
idx_min = dis.idxmin(axis = 1)
idx_min
```

```python
groups = idx_min.groupby(idx_min).groups
groups
```

```python
references
```

```python
order = []
for key in references:
    if key in groups:
        values = [key]+list(groups[key])
        order = order + values
    else:
        order = order + [key]
```

```python
order
```

```python
filename = '/Users/ktruong/workspace/Workspace/experiments/056_6_SPECIES_SELECTION_PERCENTAGE/preorders/e_coli/01/order.txt'
values = order
with open(filename, "w") as f:
    for value in values:
        f.write(f"{value}\n")
```
