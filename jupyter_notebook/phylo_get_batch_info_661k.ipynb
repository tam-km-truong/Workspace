```python
def get_nb_of_species_in_batch(batch_name):
    file = open(f"/Users/ktruong/data/661k/batches_text_file/{batch_name}.txt","r")
    genomes = file.readlines()
    genomes = [gen.replace(".fa.gz\n","") for gen in genomes]
    group_sizes = bakrep_metadata.loc[genomes].groupby(['gtdbtk.classification.species']).size()
    return len(group_sizes)
```

```python
for idx in range(1,23):
    if idx<10:
        name = f"dustbin__0{idx}"
    else:
        name = f"dustbin__{idx}"
    nb = get_nb_of_species_in_batch(name)
    print(name, ":", nb)
```
